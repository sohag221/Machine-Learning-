```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/income.csv')
```

```python
df.head()
```

```python
df.drop('Name', axis='columns', inplace=True)
```

```python
df
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(df[['Age', 'Income($)']])
```

```python
predicted = km.fit_predict(df[['Age', 'Income($)']])
```

```python
predicted
```

```python
df['cluster1'] = predicted
```

```python
df
```

```python
df1 = df[df.cluster1==0]
df2 = df[df.cluster1==1]
df3 = df[df.cluster1==2]
```

```python
plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='blue')
```

```python
scaler = MinMaxScaler()

```

```python
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df
```

```python
km = KMeans(n_clusters=3)
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
```

```python
y_predicted
```

```python
df['cluster2'] = y_predicted
```

```python
df
```

```python
df4 = df[df.cluster2==0]
df5 = df[df.cluster2==1]
df6 = df[df.cluster2==2]
```

```python
km.cluster_centers_
```

```python
plt.scatter(df4['Age'], df4['Income($)'], color='green')
plt.scatter(df5['Age'], df5['Income($)'], color='red')
plt.scatter(df6['Age'], df6['Income($)'], color='blue')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],marker='*' ,color='yellow')
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age', 'Income($)']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range, sse)
#
```
